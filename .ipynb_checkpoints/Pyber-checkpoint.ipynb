```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
city_data_to_load = 'Resources/city_data.csv'
ride_data_to_load = 'Resources/ride_data.csv'
```

```python
# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
ride_data_complete = pd.merge(city_data, ride_data, how='left', on=['city'])
```

```python
# Display the data table for preview
ride_data_complete.head()
```

```python
ride_data_grouped = ride_data_complete.groupby('city')
```

```python
average_fare = ride_data_grouped['fare'].mean()
```

```python
total_rides = ride_data_grouped['ride_id'].count()
```

```python
total_drivers = ride_data_grouped['driver_count'].max()
```

```python
city_type = ride_data_grouped['type'].max()
```

```python
bubble_plot_df = pd.DataFrame({'Average Fare ($)': average_fare,
                   'Total Rides': total_rides,
                   'Total Drivers': total_drivers, 
                   'City Type': city_type})
bubble_plot_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types

urban = bubble_plot_df.loc[bubble_plot_df['City Type'] == 'Urban']
suburban = bubble_plot_df.loc[bubble_plot_df['City Type'] == 'Suburban']
rural = bubble_plot_df.loc[bubble_plot_df['City Type'] == 'Rural']

x1 = urban['Total Rides']
y1 = urban['Average Fare ($)']

x2 = suburban['Total Rides']
y2 = suburban['Average Fare ($)']

x3 = rural['Total Rides']
y3 = rural['Average Fare ($)']

# Build the scatter plots for each city types

sizes = [urban['Total Drivers'] * 10, suburban['Total Drivers'] * 10, rural['Total Drivers'] * 10]
colors = ['coral', 'lightskyblue', 'gold']

plt.scatter(x1, y1, s=sizes[0], linewidths=1, c=colors[0], edgecolor='black', alpha=0.75, label='Urban')
plt.scatter(x2, y2, s=sizes[1], linewidths=1, c=colors[1], edgecolor='black', alpha=0.75, label='Suburban')
plt.scatter(x3, y3, s=sizes[2], linewidths=1, c=colors[2], edgecolor='black', alpha=0.75, label='Rural')

# Incorporate the other graph properties

plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

plt.xlim(0, 41)
plt.ylim(19, 44)
plt.grid(True)

# Create a legend

legend = plt.legend(title='City Types', loc='best')

for handler in legend.legendHandles:
    handler._sizes = [30]
    
# Incorporate a text label regarding circle size

plt.text(42, 35, 'Note:\nCircle size correlates with driver count per city.')

plt.show()
```

```python
ride_type_urban = ride_data_complete.loc[ride_data_complete['type'] == 'Urban']
ride_type_suburban = ride_data_complete.loc[ride_data_complete['type'] == 'Suburban']
ride_type_rural = ride_data_complete.loc[ride_data_complete['type'] == 'Rural']
```

```python
# Calculate Type Percents

urban_fare_pct = ride_type_urban['fare'].sum() / ride_data_complete['fare'].sum() * 100
suburban_fare_pct = ride_type_suburban['fare'].sum() / ride_data_complete['fare'].sum() * 100
rural_fare_pct = ride_type_rural['fare'].sum() / ride_data_complete['fare'].sum() * 100

# Build Pie Chart

labels = ['Urban', 'Rural', 'Suburban']

colors = ['coral', 'gold', 'lightskyblue']

explode = (0.1, 0, 0)

total_fare_pct = [urban_fare_pct, rural_fare_pct, suburban_fare_pct]

# Save Figure

plt.pie(total_fare_pct, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=280)

plt.title('% of Total Fares by City Type')

plt.show()
```

```python
# Calculate Ride Percents

urban_ride_pct = ride_type_urban['ride_id'].count() / ride_data_complete['ride_id'].count() * 100
suburban_ride_pct = ride_type_suburban['ride_id'].count() / ride_data_complete['ride_id'].count() * 100
rural_ride_pct = ride_type_rural['ride_id'].count() / ride_data_complete['ride_id'].count() * 100

# Build Pie Chart

labels = ['Urban', 'Rural', 'Suburban']

colors = ['coral', 'gold', 'lightskyblue']

explode = (0.1, 0, 0)

total_ride_pct = [urban_ride_pct, rural_ride_pct, suburban_ride_pct]

# Save Figure

plt.pie(total_ride_pct, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=270)

plt.title('% of Total Rides by City Type')

plt.show()
```

```python
# Calculate Type Percents

ride_type_urban_grouped = ride_type_urban.groupby('city')
ride_type_suburban_grouped = ride_type_suburban.groupby('city')
ride_type_rural_grouped = ride_type_rural.groupby('city')

urban_driver_pct = ride_type_urban_grouped['driver_count'].max().sum() / ride_data_grouped['driver_count'].max().sum() * 100
suburban_driver_pct = ride_type_suburban_grouped['driver_count'].max().sum() / ride_data_grouped['driver_count'].max().sum() * 100
rural_driver_pct = ride_type_rural_grouped['driver_count'].max().sum() / ride_data_grouped['driver_count'].max().sum() * 100

# Build Pie Chart

labels = ['Urban', 'Rural', 'Suburban']

colors = ['coral', 'gold', 'lightskyblue']

explode = (0.1, 0, 0)

total_driver_pct = [urban_driver_pct, rural_driver_pct, suburban_driver_pct]

# Save Figure

plt.pie(total_driver_pct, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=230)

plt.title('% of Total Drivers by City Type')

plt.show()
```
